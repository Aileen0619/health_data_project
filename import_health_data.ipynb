```python
import pandas as pd
```

```python
file_path = "Data/2025 County Health Rankings Data - v1.xlsx"
sheet_name = "Select Measure Data"
```

```python
# Read the full sheet while setting the multi-header structure
df = pd.read_excel(file_path, sheet_name=sheet_name, header=[0, 1])
```

```python
df
```

```python

# Concatenate multi-index columns to create unique headers
df.columns = [f"{str(col[0]).strip()} - {str(col[1]).strip()}" for col in df.columns]

# Display the first few rows to confirm structure
df.head()
```

```python
column_names = list(df.columns)
column_names[:3] = ["FIPS", "State", "County"]
df.columns = column_names
```

```python
df.columns
```

```python
import re

# Keep original column names as descriptions
column_descriptions = df.columns.tolist()

# Create cleaned column names for BigQuery
cleaned_columns = []
for col in column_descriptions:
    # Replace special characters and patterns
    col_clean = col
    col_clean = re.sub(r'\s+', '_', col_clean)             # Replace spaces with underscores
    col_clean = col_clean.replace('#', 'count')            # Replace # with 'count'
    col_clean = col_clean.replace('%', 'pct')              # Replace % with 'pct'
    col_clean = col_clean.replace('2+', '2more')           # Replace '2+' with '2more'
    col_clean = re.sub(r'[()\-,]', '_', col_clean)         # Replace (, ), - and , with underscores
    col_clean = re.sub(r'__+', '_', col_clean)             # Collapse multiple underscores
    col_clean = col_clean.strip('_')                       # Remove leading/trailing underscores
    cleaned_columns.append(col_clean)

# Assign cleaned column names
df.columns = cleaned_columns

```

```python
df
```

```python
# Display mapping of original to cleaned column names
column_mapping = pd.DataFrame({
    "Original Column Name": column_descriptions,
    "Cleaned Column Name": cleaned_columns
})
```
